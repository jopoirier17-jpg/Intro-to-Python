```python
# One of the release Year for Appolo 11 movie is PG wich stop us to do mathematical function on it. 
# 
def fix_released_year(x): # define a function 
    if x == "PG": #find the PG in Released Year
        return 1995 # Return 1995 instead of PG
    else:
        return int(x) # all other value in the column are now Interger

df["Released_Year"] = df["Released_Year"].apply(fix_released_year) # apply the funtion to ["Released_Year"]
mask = df["Released_Year"] > 2010 
df[mask]
```

```python
def fix_runtime(x):
    # Extracts the number and returns it
    if type(x) == int: #if already an int this will return x without converting
        return x
    return int(x[:-4])

# Use square brackets for column access
df["Runtime"]= df["Runtime"].apply(fix_runtime)

df
```

```python
def fix_Metascore(x):
    try:
        return int(x)
    except:
        return 0
df["Meta_score"] = df["Meta_score"].apply(fix_Metascore)
df
```
