```python
#global variable and how to set exceptions
def matrix_addition(a, b):

    #take two matrices and returns the addition result as outpout
    assert len(a) == len(b), "the number of row does not match" #check the number of row if they are equal
    assert len(a[0])==len(b[0]), "the number of colomn does not match" #check the number of column if they match
    c= [
        [0, 0],
        [0, 0]
    ]
    for i in range(len(a)):
        for j in range(len(a[i])):
            c[i][j] = a[i][j] * b[i][j]
    
    return c
```

```python
#####################################  Define the class Book #########################################################

class Book: 
    def __init__(self, title, author):  # take as arguments the title and author of the book
        self.title = title
        self.author = author
        self.__is_borrowed = False     # also define a protected attribute to save whether a book is borrowed or not

    def borrow_book(self):
        if self.__is_borrowed:
            return False
        else:
            self.__is_borrowed = True
            return True

    def return_book(self):
        self.__is_borrowed = False # just change the __is_borrowed attribute


    def is_available(self):  # print if the book is available or not

        if self.__is_borrowed:
            return "Le livre n'est pas disponible"     # __is_borrowed = True -> the book is not available
        return "Le livre est disponible"     # __is_borrowed = False -> the book is available
    
##########################################  Define the subclass Library  #############################################

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def show_inventory(self):
        print("--- Inventaire de la bibliothèque ---")
        for book in self.books:
            print(f"Titre: {book.title}, Auteur: {book.author}, Disponibilité: {book.is_available()}")
        print("-------------------------")

######################################################################################################################

# Usage
lib = Library()
b1 = Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams")
b2 = Book("Chronicles of St Marys Q", "Jodi Taylors")
b3 = Book("Lord of the Rings", "Tolkien")

lib.add_book(b1)
lib.add_book(b2)
lib.add_book(b3)

lib.show_inventory()

print("Première essaie d'emprunter le b1: ", b1.borrow_book())  # Should be True
print("Deuxième essaie d'emprunter le b1: ", b1.borrow_book()) # Should be False

lib.show_inventory()
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
 
class Library:
    # create a list of books available in the library and their counts
    def __init__(self):
        self.inventory = {}
 
    def add_book(self, book):

        if book.title in self.inventory:
        # if the book is already in the inventory, increment its count by one
            self.inventory[book.title]+=1
        
        else: # if not, add it to the inventory and set the count to one
            self.inventory[book.title]=1
        print(f"added: {book.title}")
        
 
    def borrow_book(self, book):
        # check if the count of the requested book is positive, approve the
        # borrow request, if its zero, deny it
        if self.inventory[book.title] > 0: #???
            self.inventory[book.title] -=1
            print(f"you have borrowed {book.title}")
        else:
            print (f"Sorry, there is no copy available for {book.title}")

    # increment the count of the borrowed book
    def return_book(self, book):
        if book.title in self.inventory.keys():
            self.inventory[book.title] += 1
            print(f"Returned: {book.title}")
        else:
            self.add_book(book)
 
    def show_inventory(self):
        # iterate over the inventory and only show the books with count > 0
        print ("\n ----- Current Inventory ------")
        empty = True
        for title, count in self.inventory.items():
            if count > 0:
                print(f"•{title}: {count} available")
                empty = False
        if empty:
            print("the library is currently empty")
 
 
lib = Library()
b1 = Book("Little Prince", "Antoine ...")
b2 = Book("Calculus", "Newton")
b3 = Book("lord of the ring", "tolkien")

lib.show_inventory()
lib.add_book(b1)
lib.add_book(b1)
lib.add_book(b2)
lib.show_inventory()

lib.borrow_book(b1)
lib.show_inventory()
lib.borrow_book(b1)
lib.show_inventory()
lib.borrow_book(b1)
lib.show_inventory()

lib.return_book(b1)
lib.show_inventory()
lib.return_book(b3)
lib.show_inventory()
```
