```python
def multiply_number(*args):\

    total = 3 #variable
    for num in args: # for the number in *args
        total *=num
    return total

#calling the function and give value that will be stored in *args
result = multiply_number(1, 2, 3) #will multiply total (3) by 1 and 2 and 3 3x1x2x3=18
print(result)
```

```python
#global variable
def matrix_addition(a, b):

    #take two matrices and returns the addition result as outpout
    
    c= [
        [0, 0],
        [0, 0]
    ]
    for i in range(len(a)):
        for j in range(len(a[i])):
            c[i][j] = a[i][j] * b[i][j]
    
    return c
```

```python

a = [
    [1, 2],
    [3, 4]
]
b = [
    [9, 8],
    [7, 6]
]
    
print(matrix_addition(a, b))
```

```python
class Car: #create a class Car

    def __init__ (self, color, make, model):

        self.make = make
    
        self.model = model

        self.color = color
    
    def description(self):
        
        print(f"This is a {self.color} {self.make} {self.model}.")
```

```python
my_car1 = Car("blue", "Toyota", "Corolla") #add a car to the Car class

my_car1.description()

my_car2 = Car(color="yellow", make= "Hyundai", model="Ioniq5")

my_car2.description()
```

```python
del(my_car2) #delete my_car2 from memory
```

```python
def display_info(self):

    print(f"This is a {self.color} {self.make} {self.model}")

class ElectricCar(Car):

        def __init__(self, color, make, model, battery_capacity):

            super().__init__(color, make, model)
            
            self.battery_capacity = battery_capacity

        def display_battery(self):
        
            print(f"this electric car has a {self.battery_capacity} KWh battery.")
```

```python
my_car3 = ElectricCar("blue", "tesla", "y", 79)
my_car3.description()
my_car3.display_battery()

my_car4 = ElectricCar("brown", "ford", "mustang", 70)
my_car4.description()
my_car4.display_battery()
```

```python

#copy data from one file to a new file or overwrite an existing one, must keep the same extension of file

def copy_paste(src_path, dest_path): #define a function that will copy from a source to a destination
    f=open(src_path, "rb") #rb read binary will include all type of text, video, image etc
    g=open(dest_path,"wb") #wb will write binary to the new document or overwrite.
    g.write(f.read())
    f.close() #close file
    g.close()

copy_paste(src_path="guide-utilisateur-epc.pdf", dest_path="copy2.pdf") #calling the function copy_paste and giving my source and destination
```
