```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
 
class Library:
    # create a list of books available in the library and their counts
    def __init__(self):
        self.inventory = {}
 
    def add_book(self, book):

        if book.title in self.inventory:
        # if the book is already in the inventory, increment its count by one
            self.inventory[book.title]+=1
        
        else: # if not, add it to the inventory and set the count to one
            self.inventory[book.title]=1
        print(f"added: {book.title}")
        
 
    def borrow_book(self, book):
        # check if the count of the requested book is positive, approve the
        # borrow request, if its zero, deny it
        if self.inventory[book.title] > 0: #???
            self.inventory[book.title] -=1
            print(f"you have borrowed {book.title}")
        else:
            print (f"Sorry, there is no copy available for {book.title}")

    # increment the count of the borrowed book
    def return_book(self, book):
        if book.title in self.inventory.keys():
            self.inventory[book.title] += 1
            print(f"Returned: {book.title}")
        else:
            self.add_book(book)
 
    def show_inventory(self):
        # iterate over the inventory and only show the books with count > 0
        print ("\n ----- Current Inventory ------")
        empty = True
        for title, count in self.inventory.items():
            if count > 0:
                print(f"•{title}: {count} available")
                empty = False
        if empty:
            print("the library is currently empty")
 
 
lib = Library()
b1 = Book("Little Prince", "Antoine ...")
b2 = Book("Calculus", "Newton")
b3 = Book("lord of the ring", "tolkien")
b4 = Book("lord of the ring 2", "tolkien")

lib.show_inventory()
lib.add_book(b1)
lib.add_book(b1)
lib.add_book(b2)
lib.add_book(b4)
lib.show_inventory()

# lib.borrow_book(b1)
# lib.show_inventory()
# lib.borrow_book(b1)
# lib.show_inventory()
# lib.borrow_book(b1)
# lib.show_inventory()

# lib.return_book(b1)
# lib.show_inventory()
# lib.return_book(b3)
# lib.show_inventory()
```
