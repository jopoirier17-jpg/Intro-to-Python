```python
import matplotlib as plt
import matplotlib.pyplot as plta
x=[1,2,3,4,5,6]
y1=[7,4,2,3,8,1]
y2=[3,7,8,9,2,4]
plta.title("Custom Figure")
plta.xlabel("Horizontal label")
plta.ylabel("Vertical Label")
plta.plot(x,y1, color='red', label='first y values')
plta.plot(x,y2, color='blue', label='second y values')
plta.legend(loc='upper left')
```

```python
plta.plot(x, y1)
plta.show()
```

```python
plta.bar(x,y2, color='blue', label='second y values')
plta.bar(x,y1, color='red', label='second y values')
plta.show
```

```python
plta.barh(x, y1)
plta.show()
```

```python
plta.scatter(x, y1)
plta.show()
```

```python
x=[1,2,3,4,5,6]
y=[7,4,2,3,8,1]
plta.hist2d(x, y)
plta.show()
```
